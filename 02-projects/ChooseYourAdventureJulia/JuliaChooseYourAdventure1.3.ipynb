```julia
function starting_room()
    println("What is your name?")
    name = readline() 
    
    println("Hello, " * name * ".")
    println("Explain the situation in the Starting Room.")
    println("Choices: 1. Stay in Room 2. End Game 3.Move to the Next Room")
    
    println("Enter a number: ")
    choice = readline() 
    
    if choice == "1"
        println("You chose to stay in the room")
        starting_room()
    elseif choice == "2"
        println("You chose to end the game")
        replay()
    elseif choice == "3"
        println("You chose to enter the New Room")
        game()
    else
        println("You entered an invalid input. We will re-enter the room. ")
        starting_room()
    end
end

function game()
    println("Quick. The dinosoar came alive")
    println("Choices: 1. Fight him 2. Hide 3. End game")
    
    println("Enter a number: ")
    choice = readline() 
    
    if choice == "1"
        println("You chose to fight him. You died. ")
        replay()
    elseif choice == "3"
        println("You chose to end the game")
        replay()
    elseif choice == "2"
        println("You chose to hide from him in a different room. You will now enter the room.")
        pool()
    else
        println("You entered an invalid input. We will re-enter the room. ")
        game()
    end
end

function pool()
    println("Oh no! There is a dog in the pool. But there is also a lion ready to attack you.")
    println("Choices: 1. Save the dog 2. Run away 3. End game")
    
    println("Enter a number: ")
    choice = readline() 
    
    if choice == "1"
        println("You chose to save him. The lion attacked you and you died. ")
        replay()
    elseif choice == "3"
        println("You chose to end the game")
        replay()
    elseif choice == "2"
        println("You chose to run away from the pool. You win!")
    else
        println("You entered an invalid input. We will re-enter the room. ")
        pool()
    end
end

function replay()
    #get user choice 
    println("Do you want to play again? Enter y or n (without quotes): ")
    choice = readline()
          
    # check user choices 
    if choice == "y"
        starting_room()
    elseif choice == "n"
        end_of_game()
    else
        println("Please enter a valid input.")
        replay()
    end
end
        
function end_of_game()
    println("Thanks for playing!")
end
    
starting_room()
```
