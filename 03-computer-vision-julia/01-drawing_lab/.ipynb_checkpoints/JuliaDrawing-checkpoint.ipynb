```julia
using Luxor

@draw begin
    setopacity(0.85)
    steps = 20
    gap   = 2
    for (n, ) in enumerate(range(0, step=2/steps, length=steps))
        sethue([Luxor.julia_purple,
            Luxor.julia_green,
            Luxor.julia_red,
            Luxor.julia_blue][mod1(n, 4)])
        sector(O, 50, 250 + 2n, ,  + 2/steps - deg2rad(gap), :fill)
    end
end
```

```julia
@draw circle(O, 20, :fill)

@draw circle(O, 20, :fill) 400

@draw circle(O, 20, :fill) 400 1200


@draw begin
         setline(10)
         sethue("pink")
         circle(O, 20, :fill)
      end


@draw begin
         setline(10)
         sethue("pink")
         circle(O, 20, :fill)
      end 1200 1200
```

```julia
Drawing()
background(1, 1, 1, 0)
origin()
setline(30)

setcolor("magenta")
box(BoundingBox() - 50, :stroke)

background(1, 1, 1, 0)
origin()
setline(20)

setcolor("cyan")
box(BoundingBox() - 50, :stroke)

background(1, 1, 1, 0)
origin()
setline(10)

setcolor("hot pink")
box(BoundingBox() - 80, :fill)

finish()
preview()
```
