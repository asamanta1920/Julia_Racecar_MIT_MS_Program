```julia
using Images
birdie = load("birdie.jpg")

white = RGB(1, 1, 1)
black = RGB(0, 0, 0)

h, s, v = 100, 244, 255
mask = map(color -> let hsv = HSV(color)
                      hsv.h <= h && hsv.s <= s/255 && hsv.v <= v/255
                    end,
           birdie)

mask_image = map(bool -> bool ? white : black,
                 mask)

mask_function(color, bool) = bool ? color : black
unmask_function(color, bool) = bool ? black :  color

# unmasked_image = map(unmask_function, birdie, mask)

masked_image = map(mask_function, birdie, mask)
```
